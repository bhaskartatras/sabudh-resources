Q4. Suppose we want two columns 'Gender' and 'Attrition'. Since they do not lie simultaneously, we will be using loc instead of iloc. iloc takes indexes to give you the results but loc uses labels and hence it is quite useful when you want to extract rows or columns that do not lie together

```
columns = ['Gender', 'Attrition']
gender_attrition = attrition_dataset.loc[:10,columns]
gender_attrition
```

Now if i want data at indexes 1, 148 and 1000, I will do the following

```
indices = [1, 148, 1000]
gender_attrition_specefic = gender_attrition.loc[indices,:]
gender_attrition_specefic
```

```
# You can also combine the two processes above.

indices = [1, 148, 1000]
columns = ['Gender', 'Attrition']
gender_attrition = gender_attrition.loc[indices,columns]
gender_attrition
```

## Renaming columns

Q9. Rename the column 'Attrition' to 'attrition'

```
df_2 = attrition_dataset.rename(columns={'Attrition':'attrition', 'Gender': 'male-female'})
```

```
attrition_dataset.head()
```

```
df_2.head()
```

```
attrition_dataset.rename(columns={'Attrition':'attrition'}, inplace=True)  #{Original_col: Rename_col}
attrition_dataset.head()
```

**inplace** by default has False values, i.e, if you don't mention the keyword it will not permanently rename the columns requested. This helps us to retain the original dataframe and we can assign to another variable if we want a dataframe with renamed columns.

You can use inplace=True if you are sure that you will not need those columns later since it permanently renames the requested columns requested.
